```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the age data
#data = {'age': [18, 19, 20, 20, 18, 19, 18, 20]}
#df = pd.DataFrame(data)

# Create a bar graph
plt.figure(figsize=(8, 6))
plt.bar(df['Age'].value_counts().index, df['Age'].value_counts().values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Number of 18s, 19s, and 20s')

# Add the count on top of each bar
for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, bar.get_height(), ha='center', va='bottom')

# Show the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the ratings data
#data = {'rating': [5, 4, 3, 2, 5, 4, 1, 5]}
#df = pd.DataFrame(data)

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Rate government performance out of 10 (2019-2023)?'], bins=5)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the party data
#data = {'party': ['BRS', 'BJP', 'Congress', 'NOTA', 'Not Willing to Vote', 'AAP', 'AIMIM']}
#df = pd.DataFrame(data)

# Filter out the 'NOTA' and 'Not Willing to Vote' values
filtered_df = df[df['Party'] != 'NOTA']
filtered_df = filtered_df[filtered_df['Party'] != 'Not voting']

# Create a pie chart
plt.figure(figsize=(8, 8))
filtered_df['Party'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Party Preference for Upcoming Elections (2023)')

# Show the pie chart
plt.show()
```
